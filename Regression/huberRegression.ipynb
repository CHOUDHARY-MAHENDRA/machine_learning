```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df=pd.read_csv("C:\\Users\\choud\\Desktop\\jupter_files\\weather.csv",names=['datetime_utc','conds','dewptm','fog','hail','heatindexm','hum','precipm','pressurem','rain','snow','tempm','thunder','tornado','vism','wdird','wdire','wgustm','windchillm','wspdm'],index_col=False)
```

```python
df.head(5)
```

```python
df.drop([0],inplace=True)
```

```python
df.head(5)
```

```python
df.drop(['conds','datetime_utc','windchillm','wgustm','precipm','heatindexm','wdire','wdird','thunder','tornado','fog','hail','wspdm','rain','snow' ],axis=1,inplace=True)
```

```python
df
```

```python
df["dewptm"] = df["dewptm"].astype(float)
df["hum"] = df["hum"].astype(float)
df["pressurem"] = df["pressurem"].astype(float)
df["tempm"] = df["tempm"].astype(float)
```

```python
df["pressurem"].replace(-9999,np.median(df["pressurem"]),inplace=True)
df["pressurem"].fillna((np.mean(df["pressurem"]))/2,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer
from sklearn.linear_model import HuberRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score

# sklearn.linear_model.HuberRegressor
```

```python
knn=KNNImputer(n_neighbors=10,weights='distance')
df_trans=knn.fit_transform(df)
```

```python
df_trans
```

```python
new_df=pd.DataFrame(df_trans)
new_df.rename(columns={new_df.columns[0]:'dewptm',new_df.columns[1]:'hum',new_df.columns[2]:'pressurem',new_df.columns[3]:'tempm',new_df.columns[4]:'vism'},inplace=True)
new_df.head(10)
```

```python
x=new_df.drop(["hum"],axis=1)
y=new_df["hum"]
```
